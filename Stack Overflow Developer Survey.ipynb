```python
def function1(parameter) :
    list2 = []
    for item in parameter.dropna().values :
        list1 = item.split(';')
        for val in list1 :
            if val not in list2 :
                list2.append(val)
    return(list2)
```

```python
def function2(parameter) :
    dic = {}
    for item in parameter.dropna().values :
        list1 = item.split(';')
        for val in list1 :
            if val in dic :
                dic[val] += 1
            else :
                dic[val] = 1
    return(dic)
```
